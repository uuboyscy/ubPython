# All URL

```python
'''

All url

'''
import json
from urllib import request

url_all_cities = 'https://works.ioa.tw/weather/api/all.json'
url_get_towns_by_cities = 'https://works.ioa.tw/weather/api/cates/%s.json'###
url_get_towns_information = 'https://works.ioa.tw/weather/api/towns/%s.json'###
url_get_url = 'https://works.ioa.tw/weather/api/url.json'
url_get_towns_weather = 'https://works.ioa.tw/weather/api/weathers/%s.json'###
```

```python
# Test url
import pprint
test_data_dict = json.loads(request.urlopen(url_all_cities).read().decode('utf-8'))

pprint.pprint(test_data_dict)
```

# Usage to each URL

```python
'''

Get all cities

'''

data_city = request.urlopen(url_all_cities).read().decode('utf-8')
data_city_dict = json.loads(data_city)

pprint.pprint(data_city_dict)
```

```python
# Print number of each city and town

for ct in data_city_dict:    
    print('===============')
    print('id :\t', ct['id'], ' , ', 'name :\t', ct['name'])
    
    for tn in ct['towns']:
        print('\t', 'id :\t', tn['id'], ' , ', 'name :\t', tn['name'])
    
    print('===============')
```

```python
'''

Get towns by single cityID
Take example for 桃園(8) 中壢區(81)

'''

data_towns = request.urlopen(url_get_towns_by_cities%(8)).read().decode('utf-8')
data_towns_dict = json.loads(data_towns)

pprint.pprint(data_towns_dict)
```

```python
'''

Get single town information by townID
Take example for 桃園(8) 中壢區(81)

'''

data_towns_information = request.urlopen(url_get_towns_information%(81)).read().decode('utf-8')
data_towns_information_dict = json.loads(data_towns_information)

pprint.pprint(data_towns_information_dict)
```

```python
'''

Get URL for images path directory

'''

data_img_path_dir = request.urlopen(url_get_url).read()
data_img_path_dir_dict = json.loads(data_img_path_dir)

pprint.pprint(data_img_path_dir_dict)
```

```python
'''

Get single town wheather information by townID
Take example for 桃園(8) 中壢區(81)

'''

data_weather_information = request.urlopen(url_get_towns_weather%(81)).read().decode('utf-8')
data_weather_information_dict = json.loads(data_weather_information)

pprint.pprint(data_weather_information_dict)
```

# Get town ID by name

```python
town_id_dict = {}

for ct in data_city_dict:
    for tn in ct['towns']:
        town_id_dict['%s,%s'%(ct['name'], tn['name'])] = [ct['id'], tn['id']]

pprint.pprint(town_id_dict)
```
